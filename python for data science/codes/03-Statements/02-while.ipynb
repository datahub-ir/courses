# break, continue, pass


```python
for letter in 'Python':
    if(letter == 'h'):
        break
    print('Current Letter :', letter)
```

```python
var = 10
while var > 0:              
    print('Current variable value :', var)
    var = var -1
    if(var == 5):
        break
```

```python
for letter in 'Python':
    if(letter == 'h'):
        continue
    print('Current Letter :', letter)
```

```python
var = 10
while var > 0:              
    var = var -1
    if(var == 5):
        continue
    print('Current variable value :', var)
```

```python
for letter in 'Python':
    if(letter == 'h'):
        pass
    print('Current Letter :', letter)
```

```python
x = 0

while(x < 10):
    print('x is currently: ',x)
    print(' x is still less than 10, adding 1 to x')
    x+=1
    if(x==3):
        print('Breaking because x==3')
        break
    else:
        print('continuing...')
        continue
```
